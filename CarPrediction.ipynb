**Dealing with Categorical Variables**

```python
fuel = pd.get_dummies(car['fuel-type'])
aspiration = pd.get_dummies(car['aspiration'],drop_first=True)
body_style= pd.get_dummies(car['body-style'],drop_first=True)
drive_wheels= pd.get_dummies(car['drive-wheels'],drop_first=True)
```

```python
car= pd.concat([car,aspiration,body_style,drive_wheels],axis=1)
```
